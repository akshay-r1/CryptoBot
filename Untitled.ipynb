```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./Data Directory/crypto-markets.csv',parse_dates=['date'], index_col='date')
print(data)
```

```python
response = requests.get("http://localhost:5000/parse",params={"q":"plot graph for bitcoin"})
response=response.json()
entities = response.get("entities")
intent = response.get("intent")
print("Entities:", entities)
print("Intent:", intent)
print(entities[0]['value'])
```

```python
data1 = data.loc[data['name'] == entities[0]['value'].title()]
```

```python
data1.index=pd.to_datetime(data1.index,format='%Y-%m-%d')
```

```python
plt.plot(data1['high'])
```

```python
plt.show()
```
